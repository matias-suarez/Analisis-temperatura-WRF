## Convertimos el tiempo de la EMA de ART a UTC 

```python
aws.index = aws.index + pd.Timedelta(hours=3)
aws['2019-03-17 12:00:00':'2019-06-01 16:00:00']
```

## Cargamos la info en un DataFrame de Pandas

```python
dfTmp_A = pd.DataFrame()

for index, Temp in wrf_A.iterrows():
    try:
        a = aws.loc[index]
    except KeyError as error:
        print(f"Index Error: {error}")
        continue

        
    dfTmp_A = dfTmp_A.append({'date': index,
                          'WRF_A': float(Temp[0]),
                          'AWS': float(a['Temp'])},
                          ignore_index=True)

dfTmp_A.set_index('date', inplace=True)
```

```python
dfTmp_B = pd.DataFrame()

for index, Temp in wrf_B.iterrows():
    try:
        a = aws.loc[index]
    except KeyError as error:
        print(f"Index Error: {error}")
        continue

        
    dfTmp_B = dfTmp_B.append({'date': index,
                          'WRF_B': float(Temp[0]),
                          'AWS': float(a['Temp'])},
                          ignore_index=True)
    
dfTmp_B.set_index('date', inplace=True)
```

```python
dfTmp_C = pd.DataFrame()

for index, Temp in wrf_C.iterrows():
    try:
        a = aws.loc[index]
    except KeyError as error:
        print(f"Index Error: {error}")
        continue

        
    dfTmp_C = dfTmp_C.append({'date': index,
                          'WRF_C': float(Temp[0]),
                          'AWS': float(a['Temp'])},
                          ignore_index=True)
    
dfTmp_C.set_index('date', inplace=True)
```

```python
dfTmp_D = pd.DataFrame()

for index, Temp in wrf_D.iterrows():
    try:
        a = aws.loc[index]
    except KeyError as error:
        print(f"Index Error: {error}")
        continue

        
    dfTmp_D = dfTmp_D.append({'date': index,
                          'WRF_D': float(Temp[0]),
                          'AWS': float(a['Temp'])},
                          ignore_index=True)
    
dfTmp_D.set_index('date', inplace=True)
```

## Plot $|T_{AWS}-T_{WRF}|$ vs $T_{AWS}$  (M√°ximas)

- **Definir fecha de inicio y fin**

```python
# Definir el inicio de la base de datos
inicio = datetime.strptime('2020-01-01 12:00', '%Y-%m-%d %H:%M')
fin = datetime.strptime('2020-04-01 12:00', '%Y-%m-%d %H:%M')


# Selecciono solo los datos que estan en el mes considerado
df_A = dfTmp_A[inicio:fin]
df_B = dfTmp_B[inicio:fin]
df_C = dfTmp_C[inicio:fin]
df_D = dfTmp_D[inicio:fin]

# Extraigo las T maximas diarias del WRF y AWS
daily_max_A = df_A.groupby(pd.Grouper(freq='D')).max()
daily_max_B = df_B.groupby(pd.Grouper(freq='D')).max()
daily_max_C = df_C.groupby(pd.Grouper(freq='D')).max()
daily_max_D = df_D.groupby(pd.Grouper(freq='D')).max()

fig = plt.figure(figsize=(12,15))

# Plot BIAS vs dia para la parametrizaci√≥n A
ax1 = fig.add_subplot(411)
ax1.plot(daily_max_A.index, daily_max_A['WRF_A']-daily_max_A['AWS'], color='red', label='T max')
plt.xlim([inicio,fin])
plt.hlines(0, inicio, fin, lw=3)
plt.hlines((daily_max_A['WRF_A']-daily_max_A['AWS']).mean(), inicio, fin, lw=3, color='green', label='Media')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('$T_{WRF}-T_{AWS}$', fontsize=15)
plt.xlabel('Fecha', fontsize=15)
plt.title(str(inicio)+' - '+str(fin)+' - WRF A', fontsize=15)
plt.grid()
plt.legend(fontsize=12)

# Plot BIAS vs dia para la parametrizaci√≥n B
ax2 = fig.add_subplot(412)
ax2.plot(daily_max_B.index, daily_max_B['WRF_B']-daily_max_B['AWS'], color='red', label='T max')
plt.xlim([inicio,fin])
plt.hlines(0, inicio, fin, lw=3)
plt.hlines((daily_max_B['WRF_B']-daily_max_B['AWS']).mean(), inicio, fin, lw=3, color='green', label='Media')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('$T_{WRF}-T_{AWS}$', fontsize=15)
plt.xlabel('Fecha', fontsize=15)
plt.title(str(inicio)+' - '+str(fin)+' - WRF B', fontsize=15)
plt.grid()
plt.legend(fontsize=12)

# Plot BIAS vs dia para la parametrizaci√≥n C
ax3 = fig.add_subplot(413)
ax3.plot(daily_max_C.index, daily_max_C['WRF_C']-daily_max_C['AWS'], color='red', label='T max')
plt.xlim([inicio,fin])
plt.hlines(0, inicio, fin, lw=3)
plt.hlines((daily_max_C['WRF_C']-daily_max_C['AWS']).mean(), inicio, fin, lw=3, color='green', label='Media')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('$T_{WRF}-T_{AWS}$', fontsize=15)
plt.xlabel('Fecha', fontsize=15)
plt.title(str(inicio)+' - '+str(fin)+' - WRF C', fontsize=15)
plt.grid()
plt.legend(fontsize=12)

# Plot BIAS vs dia para la parametrizaci√≥n D
ax4 = fig.add_subplot(414)
ax4.plot(daily_max_D.index, daily_max_D['WRF_D']-daily_max_D['AWS'], color='red', label='T max')
plt.xlim([inicio,fin])
plt.hlines(0, inicio, fin, lw=3)
plt.hlines((daily_max_D['WRF_D']-daily_max_D['AWS']).mean(), inicio, fin, lw=3, color='green', label='Media')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('$T_{WRF}-T_{AWS}$', fontsize=15)
plt.xlabel('Fecha', fontsize=15)
plt.title(str(inicio)+' - '+str(fin)+' - WRF D', fontsize=15)
plt.grid()
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()
```
