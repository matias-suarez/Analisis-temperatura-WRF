## Importamos algunas librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

## Cargamos los datos de las parametrizaciones A, B, C y D y la EMA

```python
wrf_A = pd.read_pickle('/home/matias/Desktop/Documents/WRF/Datos WRF/lh_wrf_A')
wrf_B = pd.read_pickle('/home/matias/Desktop/Documents/WRF/Datos WRF/lh_wrf_B')
wrf_C = pd.read_pickle('/home/matias/Desktop/Documents/WRF/Datos WRF/lh_wrf_C')
wrf_D = pd.read_pickle('/home/matias/Desktop/Documents/WRF/Datos WRF/lh_wrf_D')

aws = pd.read_pickle('/home/matias/Desktop/Documents/WRF/Datos EMAs/lh_ema')
```

## Convertimos el tiempo de la EMA de ART a UTC 

```python
aws.index = aws.index + pd.Timedelta(hours=3)
aws
```

## Cargamos la info en un DataFrame de Pandas

```python
dfTmp_A = pd.DataFrame()

for index, Temp in wrf_A.iterrows():
    try:
        a = aws.loc[index]
    except KeyError as error:
        print(f"Index Error: {error}")
        continue

        
    dfTmp_A = dfTmp_A.append({'date': index,
                          'WRF_A': float(Temp[0]),
                          'AWS': float(a['Temp'])},
                          ignore_index=True)

dfTmp_A.set_index('date', inplace=True)
```

```python
dfTmp_B = pd.DataFrame()

for index, Temp in wrf_B.iterrows():
    try:
        a = aws.loc[index]
    except KeyError as error:
        print(f"Index Error: {error}")
        continue

        
    dfTmp_B = dfTmp_B.append({'date': index,
                          'WRF_B': float(Temp[0]),
                          'AWS': float(a['Temp'])},
                          ignore_index=True)
    
dfTmp_B.set_index('date', inplace=True)
```

```python
dfTmp_C = pd.DataFrame()

for index, Temp in wrf_C.iterrows():
    try:
        a = aws.loc[index]
    except KeyError as error:
        print(f"Index Error: {error}")
        continue

        
    dfTmp_C = dfTmp_C.append({'date': index,
                          'WRF_C': float(Temp[0]),
                          'AWS': float(a['Temp'])},
                          ignore_index=True)
    
dfTmp_C.set_index('date', inplace=True)
```

```python
dfTmp_D = pd.DataFrame()

for index, Temp in wrf_D.iterrows():
    try:
        a = aws.loc[index]
    except KeyError as error:
        print(f"Index Error: {error}")
        continue

        
    dfTmp_D = dfTmp_D.append({'date': index,
                          'WRF_D': float(Temp[0]),
                          'AWS': float(a['Temp'])},
                          ignore_index=True)
    
dfTmp_D.set_index('date', inplace=True)
```

## Plot Pearson, RMSE, MAE y BIAS vs mes del año para la parametrización A

- **Verificar el rango del índice i del loop**
- **Definir la fecha de inicio de la base de datos**

```python
# Creo algunos arrays vacios para trabajar mas adelante
pearson_array_max = np.array([])
rmse_array_max = np.array([])
mae_array_max = np.array([])
bias_array_max = np.array([])
pearson_array_min = np.array([])
rmse_array_min = np.array([])
mae_array_min = np.array([])
bias_array_min = np.array([])
month_array = np.array([])

# loop para barrer sobre cada mes de la base de datos
for i in range(13): # Aca poner el número de meses a evaluar. Tener en cuenta que range(x) va desde 0 a x-1
#     #dfTmp_A.index = pd.to_datetime(dfTmp_A['date'])
    
    # Definir el inicio de la base de datos
    inicio = datetime.strptime('2019-06-01 12:00', '%Y-%m-%d %H:%M') + relativedelta(months=i)
    fin = inicio + relativedelta(months=1)
    
    # Selecciono solo los datos que estan en el mes considerado
    df_A = dfTmp_A[inicio:fin]

    # Extraigo las T maximas diarias del WRF y AWS
    daily_max_A = df_A.groupby(pd.Grouper(freq='D')).max()
    daily_min_A = df_A.groupby(pd.Grouper(freq='D')).min()
#     print(daily_max_A)

#     Plot T AWS vs T WRF
#     fig = plt.figure(figsize=(10,4))
#     ax1 = fig.add_subplot(121)
#     ax1.scatter(daily_max_A['AWS'],daily_max_A['WRF_A'],s=50, color="none", edgecolor="red",
#                 label='Pearson = '+str(round(daily_max_A.corr(method='pearson')['AWS']['WRF_A'], 3)))
#     plt.axline([0, 0], [50, 50]) # Linea 45°
#     plt.title(str(inicio)+' - '+str(fin))
#     plt.xlabel('T° AWS')
#     plt.ylabel('T° WRF (A)')
#     plt.legend()
        
#     ax2 = fig.add_subplot(122)
#     ax2.scatter(daily_min_A['AWS'],daily_min_A['WRF_A'],s=50, color="none", edgecolor="blue", 
#                 label='Pearson ='+str(round(daily_min_A.corr(method='pearson')['AWS']['WRF_A'], 3)))
#     plt.axline([-10, -10], [30, 30]) # Linea 45°
#     plt.title(str(inicio)+' - '+str(fin))
#     plt.xlabel('T° AWS')
#     plt.ylabel('T° WRF (A)')
#     plt.legend()
#     plt.show()

    N = daily_max_A['WRF_A'].count() # Calculo el numero de datos en un mes para calcular RMSE mas adelante
    N1 = daily_min_A['WRF_A'].count()
    
    # Calculo del RMSE
    RMSE_max = (((daily_max_A['WRF_A'] - daily_max_A['AWS'])**2).sum()/N)**(.5)
    #print('RMSE max =', RMSE_max,'°C')
    RMSE_min = (((daily_min_A['WRF_A'] - daily_min_A['AWS'])**2).sum()/N1)**(.5)
    #print('RMSE min =', RMSE_min,'°C')

    # Calculo de MAE
    MAE_max = np.abs(daily_max_A['WRF_A'] - daily_max_A['AWS']).sum()/N
    #print('MAE max =', MAE_max,'°C')
    MAE_min = np.abs(daily_min_A['WRF_A'] - daily_min_A['AWS']).sum()/N1
    #print('MAE min =', MAE_min,'°C')

    # Calculo BIAS
    BIAS_max = (daily_max_A['WRF_A'] - daily_max_A['AWS']).sum()/N
    #print('BIAS_max =', BIAS_max,'°C')
    BIAS_min = (daily_min_A['WRF_A'] - daily_min_A['AWS']).sum()/N1
    #print('BIAS min =', BIAS_min,'°C')

    
    # Guardo en arrays el CC de Pearson, RMSE, MAE, BIAS y meses para plotear despues
    pearson_array_max = np.append(pearson_array_max, daily_max_A.corr(method='pearson')['AWS']['WRF_A'])
    rmse_array_max = np.append(rmse_array_max, RMSE_max)
    mae_array_max = np.append(mae_array_max, MAE_max)
    bias_array_max = np.append(bias_array_max, BIAS_max)
    
    pearson_array_min = np.append(pearson_array_min, daily_min_A.corr(method='pearson')['AWS']['WRF_A'])
    rmse_array_min = np.append(rmse_array_min, RMSE_min)
    mae_array_min = np.append(mae_array_min, MAE_min)
    bias_array_min = np.append(bias_array_min, BIAS_min)
    month_array = np.append(month_array, inicio + relativedelta(days=15))

    
fig = plt.figure(figsize=(14,8))

# Plot CC de Pearson vs Mes
ax1 = fig.add_subplot(221)
ax1.plot(month_array, pearson_array_max, marker='o', color='red', label='T max')
ax1.plot(month_array, pearson_array_min, marker='o', color='blue', label='T min')
plt.xlabel('Fecha')
plt.ylabel('Pearson')
plt.title('Comparación T - WRF A')
plt.grid()
plt.legend()

# Plot RMSE vs mes
ax2 = fig.add_subplot(222)
ax2.plot(month_array, rmse_array_max, marker='o', color='red', label='T max')
ax2.plot(month_array, rmse_array_min, marker='o', color='blue', label='T min')
plt.xlabel('Fecha')
plt.ylabel('RMSE [°C]')
plt.title('Comparación T - WRF A')
plt.grid()
plt.legend()

# Plot MAE vs mes
ax3 = fig.add_subplot(223)
ax3.plot(month_array, mae_array_max, marker='o', color='red', label='T max')
ax3.plot(month_array, mae_array_min, marker='o', color='blue', label='T min')
plt.xlabel('Fecha')
plt.ylabel('MAE [°C]')
plt.title('Comparación T - WRF A')
plt.grid()
plt.legend()

# Plot BIAS vs mes
ax4 = fig.add_subplot(224)
ax4.plot(month_array, bias_array_max, marker='o', color='red', label='T max')
ax4.plot(month_array, bias_array_min, marker='o', color='blue', label='T min')
plt.ylim([-8,8])
plt.xlabel('Fecha')
plt.ylabel('BIAS [°C]')
plt.title('Comparación T - WRF A')
plt.grid()
plt.legend()


plt.tight_layout()
plt.show()
```

```python
# Creo algunos arrays vacios para trabajar mas adelante
pearson_array_max = np.array([])
rmse_array_max = np.array([])
mae_array_max = np.array([])
bias_array_max = np.array([])
pearson_array_min = np.array([])
rmse_array_min = np.array([])
mae_array_min = np.array([])
bias_array_min = np.array([])
month_array = np.array([])

# loop para barrer sobre cada mes de la base de datos
for i in range(13): # Aca poner el número de meses a evaluar. Tener en cuenta que range(x) va desde 0 a x-1
    
    # Definir el inicio de la base de datos
    inicio = datetime.strptime('2019-06-01 12:00', '%Y-%m-%d %H:%M') + relativedelta(months=i)
    fin = inicio + relativedelta(months=1)
    
    # Selecciono solo los datos que estan en el mes considerado
    df_D = dfTmp_D[inicio:fin]

    # Extraigo las T maximas diarias del WRF y AWS
    daily_max_D = df_D.groupby(pd.Grouper(freq='D')).max()
    daily_min_D = df_D.groupby(pd.Grouper(freq='D')).min()
#     print(daily_max_D)

#     Plot T AWS vs T WRF
#     fig = plt.figure(figsize=(10,4))
#     ax1 = fig.add_subplot(121)
#     ax1.scatter(daily_max_D['AWS'],daily_max_D['WRF_D'],s=50, color="none", edgecolor="red",
#                 label='Pearson = '+str(round(daily_max_D.corr(method='pearson')['AWS']['WRF_D'], 3)))
#     plt.axline([0, 0], [50, 50]) # Linea 45°
#     plt.title(str(inicio)+' - '+str(fin))
#     plt.xlabel('T° AWS')
#     plt.ylabel('T° WRF (D)')
#     plt.legend()
        
#     ax2 = fig.add_subplot(122)
#     ax2.scatter(daily_min_D['AWS'],daily_min_D['WRF_D'],s=50, color="none", edgecolor="blue", 
#                 label='Pearson ='+str(round(daily_min_D.corr(method='pearson')['AWS']['WRF_D'], 3)))
#     plt.axline([-10, -10], [30, 30]) # Linea 45°
#     plt.title(str(inicio)+' - '+str(fin))
#     plt.xlabel('T° AWS')
#     plt.ylabel('T° WRF (D)')
#     plt.legend()
#     plt.show()
    
    
    N = daily_max_D['WRF_D'].count() # Calculo el numero de datos en un mes para calcular RMSE mas adelante
    N1 = daily_min_D['WRF_D'].count()
    
    # Calculo del RMSE
    RMSE_max = (((daily_max_D['WRF_D'] - daily_max_D['AWS'])**2).sum()/N)**(.5)
    #print('RMSE max =', RMSE_max,'°C')
    RMSE_min = (((daily_min_D['WRF_D'] - daily_min_D['AWS'])**2).sum()/N1)**(.5)
    #print('RMSE min =', RMSE_min,'°C')

    # Calculo de MAE
    MAE_max = np.abs(daily_max_D['WRF_D'] - daily_max_D['AWS']).sum()/N
    #print('MAE max =', MAE_max,'°C')
    MAE_min = np.abs(daily_min_D['WRF_D'] - daily_min_D['AWS']).sum()/N1
    #print('MAE min =', MAE_min,'°C')

    # Calculo BIAS
    BIAS_max = (daily_max_D['WRF_D'] - daily_max_D['AWS']).sum()/N
    #print('BIAS_max =', BIAS_max,'°C')
    BIAS_min = (daily_min_D['WRF_D'] - daily_min_D['AWS']).sum()/N1
    #print('BIAS min =', BIAS_min,'°C')

    
    # Guardo en arrays el CC de Pearson, RMSE, MAE, BIAS y meses para plotear despues
    pearson_array_max = np.append(pearson_array_max, daily_max_D.corr(method='pearson')['AWS']['WRF_D'])
    rmse_array_max = np.append(rmse_array_max, RMSE_max)
    mae_array_max = np.append(mae_array_max, MAE_max)
    bias_array_max = np.append(bias_array_max, BIAS_max)
    
    pearson_array_min = np.append(pearson_array_min, daily_min_D.corr(method='pearson')['AWS']['WRF_D'])
    rmse_array_min = np.append(rmse_array_min, RMSE_min)
    mae_array_min = np.append(mae_array_min, MAE_min)
    bias_array_min = np.append(bias_array_min, BIAS_min)
    month_array = np.append(month_array, inicio + relativedelta(days=15))

    
fig = plt.figure(figsize=(14,8))

# Plot CC de Pearson vs Mes
ax1 = fig.add_subplot(221)
ax1.plot(month_array, pearson_array_max, marker='o', color='red', label='T max')
ax1.plot(month_array, pearson_array_min, marker='o', color='blue', label='T min')
plt.xlabel('Fecha')
plt.ylabel('Pearson')
plt.title('Comparación T - WRF D')
plt.grid()
plt.legend()

# Plot RMSE vs mes
ax2 = fig.add_subplot(222)
ax2.plot(month_array, rmse_array_max, marker='o', color='red', label='T max')
ax2.plot(month_array, rmse_array_min, marker='o', color='blue', label='T min')
plt.xlabel('Fecha')
plt.ylabel('RMSE [°C]')
plt.title('Comparación T - WRF D')
plt.grid()
plt.legend()

# Plot MAE vs mes
ax3 = fig.add_subplot(223)
ax3.plot(month_array, mae_array_max, marker='o', color='red', label='T max')
ax3.plot(month_array, mae_array_min, marker='o', color='blue', label='T min')
plt.xlabel('Fecha')
plt.ylabel('MAE [°C]')
plt.title('Comparación T - WRF D')
plt.grid()
plt.legend()

# Plot BIAS vs mes
ax4 = fig.add_subplot(224)
ax4.plot(month_array, bias_array_max, marker='o', color='red', label='T max')
ax4.plot(month_array, bias_array_min, marker='o', color='blue', label='T min')
plt.ylim([-8,8])
plt.xlabel('Fecha')
plt.ylabel('BIAS [°C]')
plt.title('Comparación T - WRF D')
plt.grid()
plt.legend()


plt.tight_layout()
plt.show()
```
